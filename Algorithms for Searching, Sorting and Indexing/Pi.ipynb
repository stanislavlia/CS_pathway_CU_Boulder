```python

def factorial(n):
    
    if n == 1 or n == 0:
        return 1
    
    return n * factorial(n-1)


def C(k,n):
    
    return factorial(n) / (factorial(n-k)*factorial(k))

factorial(100)


```

```python
import matplotlib.pyplot as plt


%matplotlib notebook

n_people = [i for i in range(1, 100)]


def P_dif_bithds(n):
    
    return (factorial(365) / factorial(365-n))/ 365**n

ps = [1 - P_dif_bithds(n) for n in n_people]

plt.plot(n_people, ps)
```
