```python
class OpenHashing():
    def __init__(self, N=4):
        self.hash_table = {}
        self.N = N
        
        #filling hash table
        for i in range(4):
            self.hash_table[str(i)] = None
            

    def simple_hash(self,s):
        sum_of_codes = 0

        for i in range(len(s)):
            sum_of_codes += ord(s[i])

        return sum_of_codes % self.N
    
    
    def insert(self, s):
        ind = self.simple_hash(s)
        
        #looking for empty slot among next inds
        for i in range(ind, self.N):
            if self.hash_table[str(i)] == None:
                self.hash_table[str(i)] = s.lower()
                return
            
        #among previous
        for i in range(ind):
            if self.hash_table[str(i)] == None:
                self.hash_table[str(i)]= s.lower()
                return
            
        self.rehash()
        self.insert(s)
            
            
            
    def rehash(self):
        self.N = self.N * 2
        
        old_values = self.hash_table.values()
        
        #update hash table
        self.hash_table = {}
        
        #filling new HT with Nones
        for i in range(self.N):
            self.hash_table[str(i)] = None
        
        
        
        
        #rehash old els
        for val in old_values:
            if val == "DELETED":
                continue
            
            self.insert(val)
            
            
    def delete(self, i):
        
        if self.hash_table[str(i)] is None:
            return
        
        
        self.hash_table[str(i)] = "DELETED"
    
    
    def find(self,s): #TO DO
        
        sup_ind = self.simple_hash(s)
        
        for i in range(sup_ind, self.N):
            if self.hash_table[str(i)] == s.lower():
                print("Found with key ", i)
                return i
            
        for i in range(sup_ind):
            if self.hash_table[str(i)] == s.lower():
                print("Found with key", i)
                return i
            
        print("NOT FOUND")
        
            
        
        
            
        
            
                
        

            
            
```
