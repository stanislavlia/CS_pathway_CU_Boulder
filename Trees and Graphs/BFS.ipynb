```python
graph = {"2" : ["3","7","5"],
          "3" : ["2","6","7"],
          "5" : ["2","8","6"],
           "6" : ["5","3"],
          "7": ["2","3","12","10"],
           "8":["5"],
           "10": ["7"],
            "12" : ["7"]}




class Queue():#FIFO
    def __init__(self):
        self.storage = []
        
    def add(self, el):
        self.storage.append(el)
        
    def pop(self):
        return self.storage.pop(0)
    
    def is_empty(self):
        return  not (self.storage)
    
    

```

```python

def bfs(graph, start):
    start = str(start)
    
    
    depths = {start : 0} #initialize table of depths
    visited = {start} #visited nodes
    parents = {start:None}
    
    q = Queue()
    q.add(start)
    
    while not q.is_empty():
        #go through neighbours
        current_v = q.pop()
        neighs = graph[current_v]
        
        for neigh in neighs:
            if neigh not in visited:
                depths[neigh] = depths[current_v] + 1 #currentv - parent for neigh
                parents[neigh] = current_v
                visited.add(neigh)
                
                q.add(neigh)
                
                
    
    result = {"depths" : depths,
              "visited" : visited,
               "parents" : parents}
    
    return result

    
    
def build_the_path(graph,start, finish):
    start  = str(start)
    finish = str(finish)
    
    bfs_result = bfs(graph,start)
    
    parents = bfs_result["parents"]
    depths = bfs_result["depths"]
    
    
    path = [finish]
    
    current_v = finish
    
    
    while parents[current_v] is not None:
        path.append(parents[current_v])
        current_v = parents[current_v]
        
        
    return "-".join(path[::-1]), depths[finish]  #path and length of way
        
    
```
