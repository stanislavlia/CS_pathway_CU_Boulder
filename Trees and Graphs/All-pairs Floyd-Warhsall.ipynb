```python
graph = {"1" : {"2":1, "3":2},
         "2" : {"4":5},
         "3" : {"2":1, "4":3, "5":7},
         "4" : {"5":2},
          "5" : {}}
```

```python
def make_ajd_matrix(graph):
    
    matrix = {}
    
    for v in graph:
        matrix[v] = {}
        
        for neigh in graph:
            if neigh in graph[v]:
                matrix[v][neigh] = graph[v][neigh]
            
            elif v == neigh:
                matrix[v][neigh] = 0
            
            else:
                matrix[v][neigh] = float("INF")
        
    return matrix


def make_next_matrix(graph):
    matrix = {}
    
    for v in graph:
        matrix[v] = {}
        
        for v2 in graph:
            if v2 in graph[v]:
                matrix[v][v2] = v2
                
            elif v == v2:
                matrix[v][v2] = v
            
            else:
                matrix[v][v2] = None
                
                
    return matrix




        



```

```python

def floyd_warshall(graph):
    #init adj matrix 
    
    distance_matrix = make_ajd_matrix(graph)
    next_matrix = make_next_matrix(graph)
    
    for k in graph:
        for i in graph:
            for j in graph:
                
                if distance_matrix[i][j] < distance_matrix[i][k] + distance_matrix[k][j]:
                    #do nothing. it is already shortest path
                    continue
                else:
                    distance_matrix[i][j] = distance_matrix[i][k] + distance_matrix[k][j]
                    
                    next_matrix[i][j] = next_matrix[i][k]
                    
    return distance_matrix, next_matrix

                
                
    
```

```python
D,N = floyd_warshall(graph)
```

## All pairs through Dijkstra runs

```python
class Que:
    def __init__(self):
        self.stor =  []
    def add(self,el):
        self.stor.append(el)
    
    def pop(self):
        return self.stor.pop(0)
    
    
def dijsktra_algorithm(graph, source):
    
    cost = {}
    parents = {}
    
    
    #initialize the table 
    for vertex in graph:
        if vertex == source:
            cost[vertex] = 0
            parents[vertex] = None
            continue
        
        cost[vertex] = float("INF")
        parents[vertex] = None
        
    
    
    que = Que()
    que.add(source)
    
    
    #Dijkstra algorithm
    
    while que.stor:
        vertex = que.pop()
        
        for neigh in graph[vertex]:
            
            #try to descrease the cost
            if cost[vertex] + graph[vertex][neigh] < cost[neigh]:
                
                cost[neigh] = cost[vertex] + graph[vertex][neigh]
                parents[neigh] = vertex
                que.add(neigh)
                
                
                
    return cost, parents
        
        
        
```

```python
def all_pairs_Dijkstra(graph):
    matrix = {}
    
    for vertex in graph:
        cost, _ = dijsktra_algorithm(graph,vertex)
        
        matrix[vertex] = cost
        
        
    return matrix

all_pairs_Dijkstra(graph)
```
