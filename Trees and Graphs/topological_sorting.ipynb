```python


adj_list = {"1" : {"4","3"},
            "2" : set(),
            "3": {"5","7",'2'},
            "4" : {"5"},
            "5" : {"6"},
            "6" : set(),
            "7" : {"6"}
           }

start = "1"

time = 1
parents = {start:None}

disc_time = {start:0}
finish_time = {}
seen = set()



def dfs_visit(graph, start): #WORKS
    global time
    if start in seen:
        return
    
    seen.add(start)
    #print(seen)
    print("We visit ", start)
    
    for n in graph[start]:
        if n not in seen:
            parents[n] = start
            disc_time[n] = time
            time += 1
            dfs_visit(graph, n)
            seen.add(n)
            
    time += 1
    finish_time[start] = time
            
            

dfs_visit(adj_list, start)
```

```python
def topological_sort(graph,start): # Works
    dfs_visit(graph,start)
    
    topl_sorted_row = sorted(finish_time, key= lambda x: finish_time[x])[::-1]
    
    return topl_sorted_row



topological_sort(adj_list,"1")
```

```python
def reverse_graph(graph): #WORKS. It changes all the directions in the directed graph
    reversed_graph = {}
    
    for v in graph:
        if graph[v] is False: #no outgoing vertexes
            continue
        for neigh in graph[v]:
            
            if neigh in reversed_graph:
                reversed_graph[neigh].add(v)
            else:
                reversed_graph[neigh] = {v}
                
    return reversed_graph


r_graph = reverse_graph(adj_list)

                


r_graph
```
