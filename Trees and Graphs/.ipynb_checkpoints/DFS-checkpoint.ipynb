```python
start = "8"

graph = {"2" : ["3","7","5"],
          "3" : ["2","6","7"],
          "5" : ["2","8","6"],
           "6" : ["5","3"],
          "7": ["2","3","12","10"],
           "8":["5"],
           "10": ["7"],
            "12" : ["7"]}



time = 1
parents = {start:None}

disc_time = {start:0}
finish_time = {}
seen = set()



def dfs_visit(graph, start): #WORKS
    global time
    if start in seen:
        return
    
    seen.add(start)
    
    for n in graph[start]:
        if n not in seen:
            parents[n] = start
            disc_time[n] = time
            time += 1
            dfs_visit(graph, n)
            seen.add(n)
            
    time += 1
    finish_time[start] = time
            
            
    
            
            
            
dfs_visit(graph, start)
```
