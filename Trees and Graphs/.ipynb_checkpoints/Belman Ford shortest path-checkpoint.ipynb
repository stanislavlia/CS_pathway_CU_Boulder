```python
graph = { "1" : {"2" : 1, "3" : 2},
          "2" : {"4" : 5},
         "3" : {"2":1, "4":3,"5":7},
          "4" : {"5" : 2},
          "5" : {}
}




for v in graph:
    for v2 in graph[v]:
        print("Edge {}-{}".format(v,v2))
```

```python


def belman_ford(graph , source):

    all_edges = []
    parents = {}
    cost = {source : 0}
    
    
    
    
    for v in graph: # read all edges and initializing
        for v2 in graph[v]:
            all_edges.append( (v, v2) )
            
        parents[v] = None
        if v not in cost:
            cost[v] = float("INF")
            

    
    #shortest path alogirtm
    
    for i in range(1, len(graph)): # N vertexes
        for v_start,v_end in all_edges:
        #relax operation
            if cost[v_start] + graph[v_start][v_end] < cost[v_end]:

                cost[v_end] = cost[v_start] + graph[v_start][v_end]
                parents[v_end] = v_start

            
            
    return cost,parents
        
    


def recover_path(cost, parents,source, finish):
    path = []
    
    curr = finish
    
    while curr != source:
        path.append(curr)
        curr = parents[curr]
        
    path.append(source)
    
    return path[::-1]


cost,parents = belman_ford(graph, "1")
    
recover_path(cost,parents,"1", "5")



def find_shortest_path(graph, source, finish):
    
    cost,parents = belman_ford(graph,source)
    path = recover_path(cost,parents,source, finish)
    
    print("Shortest path ", "->".join(path), "distance ", cost[finish])
    
    return cost,parents,path



```
