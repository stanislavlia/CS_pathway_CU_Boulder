```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
dx = 0.001
l = np.pi

x = np.arange(-1+dx, 1+dx, dx)
n = len(x)

nquart = int(np.floor(n/4))

```

```python
#functions

f = np.zeros_like(x)
    
f[nquart:2*nquart] = (4/n) * np.arange(1,nquart+1)
f[2*nquart: 3*nquart] = np.ones(nquart) - np.arange(0,nquart) * (4/n)

plt.plot(x,f)
```

```python
#FT


A0 = np.sum(f) * dx


A = np.zeros(20)
B = np.zeros(20)

fFs = A0/2

for k in range(20):
    A[k] = np.sum(f * np.cos((np.pi*2*(k+1)/l))) * dx
    B[k] = np.sum(f * np.sin((np.pi*2*(k+1)/l))) * dx
    
    fFs += A[k] * np.cos((np.pi*2*(k+1)/l) + B[k]*np.sin(np.pi*2*(k+1)/l))
    
    
```
