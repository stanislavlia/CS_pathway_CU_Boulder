```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def intensity(x):
    
    return 2*math.cos(x)



x = np.arange(0,20, 0.1)

intens =  [intensity(i) for i in x]

plt.plot(x,intens)
```

```python

def kth_root(k,freq, N):

    wn = np.exp( 2j*math.pi*freq*k)
    
    return wn


def count_poly(a_coefs, x):
    result = 0
    
    for i in range(len(a_coefs)):
        
        member = a_coefs[i] * x**i
        result += member
        
    return result
        
    



def fourierTransform(f_values):
    
    N = len(f_values)
    wns_generators = [kth_root(1, i/N, N) for i in range(N)]
    
    fourier_coeffs = []
    
    for wn in wns_generators:
        f_coef  = count_poly(f_values, wn)
        fourier_coeffs.append(f_coef)
    
    return np.round(fourier_coeffs,6)
    




    


def plot_fourier(fft_coefs):
    xs = np.arange(0,len(fft_coefs),1) / len(fft_coefs)
    
    fft_real_part = np.abs(fft_coefs)
    
    plt.plot(xs, fft_real_part)
    
    
```

```python
fft_intens = fourierTransform(intens)

plot_fourier(fft_intens)
```

# fast polynomial multiplication by FFT

```python


def poly_multiply(a_coeffs, b_coeffs):
    assert len(a_coeffs) == len(b_coeffs)
    
    
    n = len(a_coeffs)
    
    #padding
    a_c_padded = np.array(a_coeffs + [0]*(n-1))
    b_c_padded = np.array(b_coeffs + [0] * (n-1))
    
    #compute fft coefs
    fft_a = np.fft.fft(a_c_padded)
    fft_b = np.fft.fft(b_c_padded)
    
    fft_res = fft_a * fft_b
    
    product = np.round(np.real(np.fft.ifft(fft_res)),5)
    
    return product


    
    
```
